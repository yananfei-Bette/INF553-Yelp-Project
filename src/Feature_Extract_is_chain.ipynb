### Feature Extract

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Read unprocessed data (old and new datasets already merged)
df_vegas_business = pd.read_json('vegas_dataset_business.json', lines = True)
```

```python
# print
df_vegas_business.head()
```

```python
num_rows = len(df_vegas_business)
print num_rows
```

```python
# create name dic
name_dic = {}
for index, smaple in df_vegas_business.iterrows():
    name_dic[smaple["name"]] = name_dic.get(smaple["name"], 0) + 1
# name_set = set(df_vegas_business["name"])
print len(name_dic)
print set(name_dic.values())
```

```python
# create is_chain column
df_vegas_business["is_chain"] = 0
```

```python
for index, _ in df_vegas_business.iterrows():
    name = df_vegas_business["name"][index]
    if name_dic[name] > 1:
        # print "***************"
        df_vegas_business.loc[index, "is_chain"] = 1
```

```python
# df_vegas_business["is_chain"]
print set(df_vegas_business["is_chain"])
```
