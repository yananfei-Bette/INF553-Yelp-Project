## age

```python
df_age = pd.read_csv("business_age.csv")
```

```python
df_age.head()
```

```python
print len(df) == len(df_age)
print len(df), len(df_age)
```

```python
# create age dic
age_dic = {}
for index, row in df_age.iterrows():
    if row[0] not in age_dic:
        age_dic[row[0]] = row[1]
    else:
        print "*******************"
        print row[0], row[1], age_dic[row[0]]
```

```python
# create new col
df["age"] = 0
```

```python
for index, row in df.iterrows():
    business_name = df["business_id"][index]
    if business_name in age_dic:
        df.loc[index, "age"] = age_dic[business_name]
    else:
        print "******************"
        print business_name
```

```python
df.head()
```

```python
# save as a json file
df.to_json("vegas_buiness_with_chain_cluster_age.json", orient = 'records', lines = True)
```
